```python
import pandas as pd
import matplotlib.pyplot as plt
trade_fairs= pd.read_csv(
    open('data-3895-2019-11-30.csv', 'r', encoding='cp1251'),
    sep=';'
)
trade_fairs.head()
```

```python
t1_count = trade_fairs.groupby(['District', 'PeriodOfPlacement']).size().sort_values(ascending=False).reset_index()
t2_count = trade_fairs.groupby(['AdmArea']).size().reset_index()
```

```python
import matplotlib
matplotlib.style.use('ggplot')

t2_count.set_index('AdmArea').plot()
plt.xticks(rotation='vertical')
```
